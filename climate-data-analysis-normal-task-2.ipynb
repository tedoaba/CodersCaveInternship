# 3. Data Cleaning

```python
# Remove suspect values
daily = daily.replace('s', pd.NA)
hourly = hourly.replace('s', pd.NA)
monthly = monthly.replace('s', pd.NA)
three_hour = three_hour.replace('s', pd.NA)
```

```python
# Remove trace precipitation amounts
daily = daily.replace('T', pd.NA)
hourly = hourly.replace('T', pd.NA)
monthly = monthly.replace('T', pd.NA)
three_hour = three_hour.replace('T', pd.NA)
```

```python
# Remove missing values
daily = daily.dropna()
hourly = hourly.dropna()
monthly = monthly.dropna()
three_hour = three_hour.dropna()
```

```python
# Remove variable wind directions
daily = daily[daily['DailyPeakWindDirection'] != 'VRB']
hourly = hourly[hourly['HourlyWindDirection'] != 'VRB']
three_hour = three_hour[three_hour['HourlyWindDirection'] != 'VRB']
```

# 4. Data Preprocessing

```python
# Preprocess the daily data
daily = daily.dropna()
daily = daily[daily['REPORT_TYPE'].str.strip() == 'SOD']

# Preprocess the hourly data
hourly = hourly.dropna()
hourly = hourly[hourly['REPORT_TYPE'].str.strip() == 'FM-15']

# Preprocess the monthly data
monthly = monthly.dropna()
monthly = monthly[monthly['REPORT_TYPE'].str.strip() == 'SOM']

# Preprocess the three-hour data
three_hour = three_hour.dropna()
three_hour = three_hour[three_hour['REPORT_TYPE'].str.strip() == 'FM-12']
```

# 5. Exploratory Data Analysis - EDA

## 5.2 Relationship between variables

```python
# Scatter plot of temperature and precipitation in monthly data
plt.figure(figsize=(8, 6))
sns.scatterplot(x='MonthlyMeanTemperature', y='MonthlyTotalLiquidPrecipitation', data=monthly)
plt.title('Relationship between Monthly Mean Temperature and Total Precipitation')
plt.xlabel('Mean Temperature (Â°F)')
plt.ylabel('Total Precipitation (inches)')
plt.show()
```

## 5.5 Monthly Aggregation

```python
# Load the daily dataset
daily = pd.read_csv('/kaggle/input/temperature-data-albany-new-york/daily_data.csv', parse_dates=['DATE'], na_values=['s', 'T', 'M'])
# Create a new column for the month
daily['Month'] = daily['DATE'].dt.month

# Calculate average temperature and precipitation per month
monthly_aggregations = daily.groupby('Month').agg({'DailyAverageDryBulbTemperature': 'mean', 'DailyPrecipitation': 'sum'})

# Display monthly aggregations
print("Monthly Aggregations:")
print(monthly_aggregations)
```

## 6.10 Seasonal Average Temprature

```python
# Extract the season from the date (e.g., Winter, Spring, Summer, Fall)
daily['Season'] = (daily['DATE'].dt.month % 12 + 3) // 3

# Calculate average temperature per season
seasonal_temperature = daily.groupby('Season')['DailyAverageDryBulbTemperature'].mean()

# Visualize trends in seasonal average temperature
plt.figure(figsize=(12, 6))
plt.plot(seasonal_temperature.index, seasonal_temperature.values, marker='o')
plt.title('Seasonal Average Temperature Trends')
plt.xlabel('Season')
plt.ylabel('Average Temperature (Â°F)')
plt.xticks([1, 2, 3, 4], ['Winter', 'Spring', 'Summer', 'Fall'])
plt.show()
```

## 6.11 Average Monthly Wind Speed

# 7. Dataset Merging

## 7.1 Select Common Columns

```python
# Load the datasets
daily_data = pd.read_csv('/kaggle/input/temperature-data-albany-new-york/daily_data.csv', na_values=['SOD  ', 'PRECIP', 'other_non_numeric_values'])
hourly_data = pd.read_csv('/kaggle/input/temperature-data-albany-new-york/hourly_data.csv', na_values=['SOD  ', 'PRECIP', 'other_non_numeric_values'])
monthly_data = pd.read_csv('/kaggle/input/temperature-data-albany-new-york/monthly_data.csv', na_values=['SOD  ', 'PRECIP', 'other_non_numeric_values'])
three_hour_data = pd.read_csv('/kaggle/input/temperature-data-albany-new-york/three_hour_data.csv', na_values=['SOD  ', 'PRECIP', 'other_non_numeric_values'])

# Identify common columns among datasets
common_columns = set(daily_data.columns) & set(hourly_data.columns) & set(monthly_data.columns) & set(three_hour_data.columns)

# Print common columns
print("Common columns:", common_columns)
```

## 7.2 Merge dataset based on selected columns

```python
# Load the datasets
daily_data = pd.read_csv('/kaggle/input/temperature-data-albany-new-york/daily_data.csv', parse_dates=['DATE'], na_values=['SOD  ', 'PRECIP', 'other_non_numeric_values'])
hourly_data = pd.read_csv('/kaggle/input/temperature-data-albany-new-york/hourly_data.csv', parse_dates=['DATE'], na_values=['SOD  ', 'PRECIP', 'other_non_numeric_values'])
monthly_data = pd.read_csv('/kaggle/input/temperature-data-albany-new-york/monthly_data.csv', parse_dates=['DATE'], na_values=['SOD  ', 'PRECIP', 'other_non_numeric_values'])
three_hour_data = pd.read_csv('/kaggle/input/temperature-data-albany-new-york/three_hour_data.csv', parse_dates=['DATE'], na_values=['SOD  ', 'PRECIP', 'other_non_numeric_values'])

# Merge the datasets based on 'DATE' with different suffixes
merged_data = pd.merge(daily_data, hourly_data, on='DATE', how='outer', suffixes=('_daily', '_hourly'))
merged_data = pd.merge(merged_data, monthly_data, on='DATE', how='outer', suffixes=('_merged', '_monthly'))
merged_data = pd.merge(merged_data, three_hour_data, on='DATE', how='outer', suffixes=('_merged', '_three_hour'))

# Display the merged dataset
print("Merged Dataset:")
merged_data.head()
```

```python
# Identify columns with mixed types in hourly_data
mixed_type_columns = hourly_data.iloc[:, [11, 12]].columns
print("Columns with mixed types in hourly_data:", mixed_type_columns)

# Print data types of those columns
print("Data types of columns with mixed types:")
print(hourly_data[mixed_type_columns].dtypes)
```

```python
# Check for missing values
missing_values = merged_data.isnull().sum()
print("Missing Values:")
print(missing_values)
```

## 7.4 Distribution of Weather Report Type

```python
# Replace 'REPORT_TYPE' with the actual column name in your datasets
report_types_daily = daily_data['REPORT_TYPE'].value_counts()
report_types_hourly = hourly_data['REPORT_TYPE'].value_counts()
report_types_three_hour = three_hour_data['REPORT_TYPE'].value_counts()

# Combine counts from all datasets
combined_report_types = report_types_daily.add(report_types_hourly, fill_value=0)
combined_report_types = combined_report_types.add(report_types_three_hour, fill_value=0)

# Plot the pie chart with previous colors
plt.figure(figsize=(10, 6))
colors = plt.cm.tab10.colors  # Using tab10 colormap for consistent colors
combined_report_types.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=colors)

# Add a legend
plt.legend(combined_report_types.index, title='Weather Report Types', loc='center left', bbox_to_anchor=(1, 0.5))

plt.title('Distribution of Weather Report Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
